```python
#Stemming

import pandas as pd
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

# Sample data for illustration
data = {'C2': ["jumps running easily cats", "playing reading books"]}
df = pd.DataFrame(data)

# Apply stemming to the words in the specified column
def stem_sentence(sentence):
    return ' '.join([stemmer.stem(word) for word in sentence.split()])

df['stemmed_column2'] = df['C2'].apply(stem_sentence)

# Display the dataframe with the new stemmed column
print(df[['C2', 'stemmed_column2']])

```

```python
#POS Tags
sentences = []

# Tokenize each sentence in the second column
for text in df['C2']:
    if pd.notnull(text):  # Check if the text is not NaN
        text_sentences = sent_tokenize(text)
        sentences.extend(text_sentences)

# Print the tokenized sentences and words with POS tags
print("\nTokenized Sentences, Words, and POS Tags from Column2:")

for sentence in sentences:
    words = word_tokenize(sentence)
    pos_tags = pos_tag(words)
    print(f"Sentence: {sentence}")
    print(f"Words: {words}")
    print(f"POS Tags: {pos_tags}")
```

```python
#Named Entity Recognition

# Initialize an empty list to store sentences
sentences = []

# Tokenize each sentence in the second column
for text in df['C2']:
    if pd.notnull(text):  # Check if the text is not NaN
        text_sentences = sent_tokenize(text)
        sentences.extend(text_sentences)

# Print the tokenized sentences, words, POS tags, and NER
print("\nTokenized Sentences, Words, POS Tags, and NER from Column2:")

for sentence in sentences:
    # Tokenize words and POS tagging
    words = word_tokenize(sentence)
    pos_tags = pos_tag(words)

    # Perform Named Entity Recognition (NER) using SpaCy
    doc = nlp(sentence)
    entities = [(entity.text, entity.label_) for entity in doc.ents]

    print(f"Sentence: {sentence}")
    print(f"Words: {words}")
    print(f"POS Tags: {pos_tags}")
    print(f"Named Entities: {entities}")

# Print a summary of the entities
print("\nSummary of Named Entities:")
all_entities = [entity for sentence in sentences for entity in nlp(sentence).ents]
entity_summary = [(entity.text, entity.label_) for entity in all_entities]
print(entity_summary)
```

```python
encoded_data_train = tokenizer.batch_encode_plus(
    df[df.data_type=='train'].C2.values,
    add_special_tokens=True,
    return_attention_mask=True,
    pad_to_max_length=True, #
    max_length=256,         #model takes 256 input strings
    return_tensors='pt'     #once everthng done it returns tensor(array in pytorch)
)

encoded_data_val = tokenizer.batch_encode_plus(
    df[df.data_type=='val'].C2.values,
    add_special_tokens=True,
    return_attention_mask=True,
    pad_to_max_length=True,
    max_length=256,
    return_tensors='pt'
)






```

SETTING UP BERT PRE TRAINED MODEL

```python
df['label'].unique()
```

```python
possible_labels = df.label.unique()
label_dict = {}
for index, possible_label in enumerate(possible_labels):
    label_dict[possible_label] = index

print(label_dict)
```

```python
from transformers import BertForSequenceClassification
model = BertForSequenceClassification.from_pretrained("bert-base-uncased",
num_labels=len(df['label'].unique()),
output_attentions=False,
output_hidden_states=False)
```

```python
from torch.utils.data import DataLoader, RandomSampler, SequentialSampler

```

```python
batch_size = 32
dataloader_train = DataLoader(dataset_train,
                              sampler=RandomSampler(dataset_train),
                              batch_size=batch_size)
dataloader_validation = DataLoader(dataset_val, sampler=SequentialSampler(dataset_val),
                                    batch_size=batch_size)
```

SETTING UP OPTIMIZER AND SCHEDULER

```python
from transformers import AdamW, get_linear_schedule_with_warmup


```

```python
optimizer = AdamW(model.parameters(),
                  lr=1e-5,
                  eps=1e-8)
```

```python
epochs = 5

scheduler = get_linear_schedule_with_warmup(optimizer,
                                            num_warmup_steps=0,
                                            num_training_steps=len(dataloader_train)*epochs)
```

8) DEFINING OUR PERFORMANCE METRICS

```python
import numpy as np
```

```python
from sklearn.metrics import f1_score
```

```python
def f1_score_func(preds, labels):
    preds_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return f1_score(labels_flat, preds_flat, average='weighted')
```

```python
def accuracy_per_class(preds, labels):
    label_dict_inverse = {v: k for k, v in label_dict.items()}

    preds_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()

    for label in np.unique(labels_flat):
        y_preds = preds_flat[labels_flat==label]
        y_true = labels_flat[labels_flat==label]
        print(f'Class: {label_dict_inverse[label]}')
        print(f'Accuracy: {len(y_preds[y_preds==label])}/{len(y_true)}\n')


```

```python
def evaluate(dataloader_val):


    model.eval()

    loss_val_total = 0
    predictions, true_vals = [], []

    for batch in dataloader_val:

        batch = tuple(b.to(device) for b in batch)

        inputs = {'input_ids':      batch[0],
                  'attention_mask': batch[1],
                  'labels':         batch[2],
                 }

        with torch.no_grad():
            outputs = model(**inputs)

        loss = outputs[0]
        logits = outputs[1]
        loss_val_total += loss.item()


        logits = logits.detach().cpu().numpy()
        label_ids = inputs['labels'].cpu().numpy()
        predictions.append(logits)
        true_vals.append(label_ids)

    loss_val_avg = loss_val_total/len(dataloader_val)
    predictions = np.concatenate(predictions, axis=0)
    true_vals = np.concatenate(true_vals, axis=0)

    return loss_val_avg, predictions, true_vals




```

```python
from tqdm.notebook import tqdm
for epoch in tqdm(range(1, epochs+1)):

    model.train()

    loss_train_total = 0

    progress_bar = tqdm(dataloader_train, desc='Epoch {:1d}'.format(epoch), leave=False, disable=False)
    for batch in progress_bar: #256/32 = 8 -- 40

        model.zero_grad()

        batch = tuple(b.to(device) for b in batch)

        inputs = {'input_ids':      batch[0],
                  'attention_mask': batch[1],
                  'labels':         batch[2],
                 }
        outputs = model(**inputs)

        loss = outputs[0]
        loss_train_total += loss.item()
        loss.backward()


        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)


        optimizer.step()
        scheduler.step()

        progress_bar.set_postfix({'training_loss': '{:.3f}'.format(loss.item()/len(batch))})

    torch.save(model.state_dict(), f'finetuned_BERT_epoch_{epoch}.model')
    tqdm.write(f'\nEpoch {epoch}')
    loss_train_avg = loss_train_total/len(dataloader_train)
    tqdm.write(f'Training loss: {loss_train_avg}')


    val_loss, predictions, true_vals = evaluate(dataloader_validation)
    val_f1 = f1_score_func(predictions, true_vals)
    tqdm.write(f'Validation loss: {val_loss}')
    tqdm.write(f'F1 Score (Weighted): {val_f1}')
```
